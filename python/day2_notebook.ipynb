```python
import numpy as np
import pandas as pd
```

```python


data = pd.read_csv('..\\data\\day2_input.txt', sep = " ", header = None, names = ['direction', 'step'])



horizontal = 0
depth = 0
aim = 0

for index,motion in data.iterrows():
    if motion['direction'] == 'forward':
        horizontal = horizontal + motion['step']
        depth = depth + aim * motion['step']
        print (horizontal, depth)
    if motion['direction'] == 'down':
        #depth = depth + motion['step']
        aim = aim + motion['step']
    if motion['direction'] == 'up':
        #depth = depth - motion['step']
        aim = aim - motion['step']

print (horizontal * depth)
```
